```python
import pandas as pd

# Открываем файл
df = pd.read_csv('parsed_data00.csv')

# Добавляем колонки
df = df.assign(Форма='Очная', Год=2)

# Проверяем изменения
print(df.head())
```

```python
# Открываем файл
df1 = pd.read_csv('parsed_data01.csv')

# Добавляем колонки
df1 = df1.assign(Форма='Очно-заочная', Год=2)

# Проверяем изменения
print(df1.head())
```

```python
# Открываем файл
df2 = pd.read_csv('parsed_data02.csv')

# Добавляем колонки
df2 = df2.assign(Форма='Заочная', Год=2.5)

# Проверяем изменения
print(df2.head())
```

```python
data = pd.concat([df, df1, df2]).reset_index(drop=True)
```

```python
data['Бюджетные'] = data['Бюджетные'].str.replace('бюджетных', '0')
```

```python
data = data.dropna(subset=['Стоимость обучения'])
```

```python
data
```

```python
data = data.assign(ВУЗ = "РАНХиГС", Балл_Б = 80, Балл_П = 80 , Макс_балл = 200)
```

```python
data = data.rename(columns={'Название программы': 'Программа'})
data = data.rename(columns={'Направление подготовки': 'Направление_подготовки_1'})
data = data.rename(columns={'Стоимость обучения': 'Стоимость'})
```

```python
data
```

```python
import pandas as pd

# предположим, что ваш датасет называется df
import pandas as pd

# Convert the values in the "Стоимость обучения" column to numeric
import pandas as pd

# Assuming your dataframe is called 'df'
data["Стоимость"] = data["Стоимость"].replace(r'\xa0', '', regex=True).values.astype(int)

data['Бюджетные'] = pd.to_numeric(data['Бюджетные'], errors='coerce')
data['Платные'] = pd.to_numeric(data['Платные'], errors='coerce')
```
